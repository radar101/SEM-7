```python
def function(x):
  return (x + 3) ** 2
```

```python
def derivative(x):
  return 2 * (x + 3)
```

```python
def gradient_descent(starting_x, learning_rate, num_iterations):
  x = starting_x
  history = [x] # to store the values values of x for each iteration

  for i in range(num_iterations):
    grad = derivative(x)
    x = x - learning_rate * grad
    history.append(x)
    # Optionally print progress every 10 iterations
    if i % 10 == 0:
        print(f"Iteration {i}: x = {x}, f(x) = {function(x)}")
  return x, history
```

```python
starting_x = 2
learning_rate = 0.01
num_iterations = 200

min_x, history = gradient_descent(starting_x, learning_rate, num_iterations)
print(f"\nThe local minimum occurs at x = {min_x}")
```

```python
# Generate points for the function
x_vals = np.linspace(-5, 5, 100)
y_vals = function(x_vals)

plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, label='y=(x+3)^2', color='blue')
plt.scatter(history, function(np.array(history)), color='red', label='Gradient Descent Path')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('Gradient Descent to Find Minimum of y=(x+3)^2')
plt.show()
```
